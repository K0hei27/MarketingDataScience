- columns：列のラベルを表します。今回のデータだと、に相当します。

## 3. 抽出したデータを集計してみる。

集計したいデータ.集計方法 とすることで、集計出来る。

## 3.1. 平均を出してみる。

平均を出す時は、.mean()を使用する。

年齢の平均値を出してみる。

```python
df.Age.mean()
```

### 【Try！】購買回数（Frequency）の平均値を出してみましょう！

## 3.2. 合計を出してみる。

個数を数える時は、.sum()を使用する。

購買回数（Frequency）の合計値を出してみる。

```python
df.Frequency.sum()
```

### 【Try！】住所（Address）の合計値を出してみましょう！文字を合計するとどうなるでしょうか？

```python
df.Address.sum()
```

## 4. データをグループごとに集計する。

性別ごとに集計してみましょう！
グループごとに同じ操作を加えたい時は、groupby().操作名というコードを使います。


今回は性別ごとの平均値を出してみましょう！

```python
df.groupby('Sex').mean()
```

ここから、性別ごとの平均年齢を抽出したい場合は…

```python
df.groupby('Sex').mean().Age
```

### 【Try！】住所（Address）ごとの合計値を出してみましょう。

このgroupbyメソッドは、月ごとの売上や購入頻度の平均、合計値等の情報が欲しい！時にも使います。
