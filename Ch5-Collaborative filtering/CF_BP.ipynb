### 0.3 レコメンドシステムの概要
#### やりたいことは、手元のデータをもとに、レコメンドしたい相手のデータの欠損値を予測すること。

ECサイト(商品のレコメンド)の場合
- 手元のデータ
    - 購買履歴データ(どのような人がどういう商品を購入してきたか)
- レコメンドしたい相手のデータ
    - その人が購入したことのある商品のデータ
    - その人が購入していない商品(欠損値)の中で一番購入する確率が高そうな商品を予測する必要がある


## 1. 簡単な例で試してみる<a class="anchor" id="first-bullet"></a>

````python
# 必要なライブラリを一括で読み込み
import pandas as pd
from numpy import nan as NA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches
from sklearn import neighbors

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 50)
````

Section 1で扱うデータを作成します。ユーザー5人（Takutoさん, Koheiさん, Jouさん, Hajimeさん, Kojiさん）が5つの映画作品に対してレビューしたデータになっています。DataFrameの作り方については[こちら](https://pythondatascience.plavox.info/pandas/pandas%E3%81%A7%E3%83%87%E3%83%BC%E3%82%BF%E3%83%95%E3%83%AC%E3%83%BC%E3%83%A0%E3%82%92%E4%BD%9C%E3%81%A3%E3%81%A6%E3%81%BF%E3%82%88%E3%81%86)を参考にしてください。

````python
df = pd.DataFrame({'name':['Takuto','Kohei','Jou','Hajime','Koji'],
                   '作品1' : np.array([5,3,4,3,1]),
                   '作品2' : np.array([3,1,3,3,5]),
                   '作品3' : np.array([4,2,4,1,5]),
                   '作品4' : np.array([2,3,2,5,2]),
                   'ダークナイト' : np.array([np.NaN,3,5,4,1])})
df = df.set_index('name', drop=True)
df.head()
````

### 1.1. イントロ
映画のレコメンドを考えてみましょう。今、手元に5つの映画の評価データがあります。Takutoさんは、4つの映画に関しては観たことがあるが、映画「[ダークナイト](https://eiga.com/movie/53214/)」はまだ観たことがありません。そこで、映画「ダークナイト」に対するTakutoさんの評価を予測して、Takutoさんの好みに合うかどうか予測してみようと思います。

イメージとしては、Takutoさんと趣味嗜好が似ている人たちの評価をベースに予測します。（ユーザーベース）つまり、2つのタスクがあります。
1. Takutoさんと他の人たちがどの程度趣味嗜好が似ているのかを定量的に評価する。
2. 似ている度合いが高い人たちの評価を参考にTakutoさんの評価を予測する。

※元ネタは[こちら](https://blog.brainpad.co.jp/entry/2017/02/03/153000)です。Pythonに書き直しました。

ヒートマップにしてみました。Takutoさんと似ているユーザーはどなたでしょう？？

````python
fig, ax = plt.subplots(figsize=(7, 5))# 見た目を整える
sns.heatmap(df, annot=True, cmap='Oranges')
ax.set_ylim(len(df), 0) # 見た目を整える
````

### 1.2. ユークリッド距離
どの程度似ているのかを定量的に評価してみます！今回の大事なポイントです。評価の仕方は複数ありますが、ここではまずユークリッド距離で評価してみましょう。
※データの類似度の評価の仕方に関しては、[この資料](https://www.slideshare.net/SeiichiUchida/21-77833992)が分かりやすかったです。


早速、TakutoさんJouさんのユークリッド距離を計算してみましょう！まずは手計算で算出してみてください。その後、Pythonで計算していきます。

````python
# Takutoさんの作品1の評価をデータフレームから抽出
df.iloc[0, 0]
# Jouさんの作品1の評価をデータフレームから抽出
df.iloc[2, 0]
# 作品1に関するユークリッド距離は
dist = (df.iloc[0, 0]-df.iloc[2, 0])**2
````

作品2から作品4まで計算して和をとってみましょう。

````python
for i in range(1, 4) :
    dist_i = (df.iloc[0, i]-df.iloc[2, i])**2
    dist = dist + dist_i

print("TakutoさんとJouさんのユークリッド距離: ",dist)
````

行列を用いると、もっと楽に早く計算することが可能に！データを扱うなら行列を学ぶべきです！自分も最近実感しました。

````python
# 行列で計算
takuto = df.iloc[0, 0:4].values
print("Takutoさんのレビュー",takuto,sep=":")
jou = df.iloc[2, 0:4].values
print("Jouさんのレビュー: ",jou)

# 距離計算
d_jou = np.sqrt(np.sum((takuto-jou)**2))
print("TakutoさんとJouさんのユークリッド距離: ",d_jou)

# コンパクトな関数もある 
np.linalg.norm(takuto - jou)
````

これを全員分やると、、、

````python
name = df.index
for i in range(5):
    print(name[i],np.linalg.norm(takuto - df.iloc[i, 0:4].values), sep = ":")
````

ユークリッド距離をデータフレームに新しい変数として結合します。

````python
# 新しい列を作成する
df["eudist"] = [0.,0.,0.,0.,0.]
for i in range(5):
    df.iloc[i,5] = np.linalg.norm(takuto - df.iloc[i, 0:4].values)
df.head()                       
````

このユークリッド距離を可視化してみます。こんなイメージです。Takutoさんと距離が近ければ近いほど似ている度合いが高いです。

````python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()
colorlist = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628', '#f781bf']

plt.plot(df.iloc[0,5],0,"o", color = colorlist[0])
plt.text(df.iloc[0,5],0.1, name[0], color = colorlist[0], size = 15)
    
for i in range(1,5):
    c = patches.Circle(xy=(0, 0), radius= df.iloc[i,5], fill=False, linestyle = "dashed", ec = colorlist[i])
    ax.add_patch(c)
    plt.plot((-1)**i*df.iloc[i,5],0,"o", color = colorlist[i])
    plt.text((-1)**i*df.iloc[i,5],0.1, name[i], color = colorlist[i], size = 15)

ax.axis("off")
````

### 1.3. 予測
さて、類似度が出ましたので、この中からTakutoさんと類似度の高いユーザを2人(JouさんとKoheiさん)選んで、ダークナイトに対する評価を推定します。類似度の高い人の評価をより重視するために、類似度を重みとして2人の平均評価値（加重平均）をとります。イメージは以下の通りです。


````python
# 画像の式をPython上で計算する
# 重みの列
df["eudist_re"] = 1/df["eudist"]
sum(df.iloc[[1,2],4].values*df.iloc[[1,2],6].values)/sum(df.iloc[[1,2],6].values)
````

### 1.4. kNN回帰
今までやってきたことをライブラリを使ってやってみましょう。

- 目的
    - Takutoさんの「ダークナイト」に対する評価を予測する
- 学習データ
    - Kohei, Jou, Hajime, Kojiさんの映画の評価
- テストデータ
    - Takutoさんの映画の評価
- ポイント
    - よりTakutoさんと似ている人の評価を重視したい!
    - kNN回帰を用いる

````python
# データは行列の形にする
df_matrix = df.values
# 学習用データ
x_train = df_matrix[1:5, 0:4]
y_train = df_matrix[1:5, 4]

# テストデータ
x_test = df_matrix[0, 0:4]

# モデル構築
# ()の中でTakutoさんと類似度の高い2人の評価を予測に使う。類似度はユークリッド距離を用いることを指定
knn2 = neighbors.KNeighborsRegressor(2, weights="distance", metric="euclidean") 
model2 = knn2.fit(x_train, y_train)

# 予測
model2.predict([x_test])
````

# 2. BigDataでトライしてみる<a class="anchor" id="second-bullet"></a>
## 2.1. 前処理
### 2.1.1. データの読み込みと加工
データソースは[こちら](https://grouplens.org/datasets/movielens/)のLatest small dataになります。誰が、どんな映画に対して、どのような評価（1～5段階で0.5刻み）をしたかを記録したデータです。レコメンドシステムの研究でベンチマークとしてよく使われるデータらしいです。データの中身はダミーではなく、ホンモノです。登録が必要になりますが、みなさんのレビューデータを追加することができるはず。

````python
df_rating = pd.read_csv('ratings.csv')
df_rating.head()
````

````python
df_movie = pd.read_csv('movies.csv')
df_movie.head()
````

movieIdで結合します。

````python
df_merged = pd.merge(df_rating, df_movie, on = "movieId")
df_merged.head()
````

Chapter1の時と同様の形にデータを加工する。インデックスがユーザーで、カラムが映画になるようにする。

````python
df_pivot = pd.pivot_table(df_merged, index = "userId", columns= "title", values = "rating")
df_pivot.head()
````

欠損多いなぁ、一人当たりどんくらい見てるのか可視化しよう。

````python
df_rating.groupby('userId').size().describe()
````

### 2.1.2. テストデータの作成
このChapterでは、自分自身へのレコメンドシステムを作成することが目的なので、みなさんの映画の評価がテストデータになります。

うる覚えタイトル検索エンジン。正確なタイトルが必要なので、、

````python
def searchmovie(name):
    print(df_movie[df_movie.title.str.contains(name,case=False)].title.unique())
````

````python
searchmovie("about time")
````

テストデータ、つまりレコメンドを出したい人の映画レビューを設定する。今回は自分で映画のレビューを10作品くらいしてみてください！

````python
# 例えば…
title = ['Titanic (1997)', 'Jaws: The Revenge (1987)',
         'Harry Potter and the Prisoner of Azkaban (2004)', 'Notebook, The (2004)', 
         'Home Alone (1990)','Aladdin (1992)',
         'Ilsa, She Wolf of the SS (1974)' 'Wolf of Wall Street, The (2013)', "Intouchables (2011)",
         'About Time (2013)', 'Léon: The Professional (a.k.a. The Professional) (Léon) (1994)']
score = [[3.6, 2.2, 3.0, 3.8, 4.0, 3.7, 5, 4.6, 4.5, 5.0]]
index = ["MY REVIEW"]
df_test = pd.DataFrame(data = score, index = index ,columns = title)
df_test.head()
````

もとのデータに結合しましょう

````python
df_test_new = pd.concat([df_pivot, df_test], axis = 0, sort=False)
df_test_new.tail()
````

欠損値は0にしておきます。(あとで欠損値のままだとモデルに通せないので…)

````python
df_test_new = df_test_new.fillna(0)
df_test_new.tail()
````

### 2.2. kNNで予測
自分がレビューした映画をもとに、自分に似ているユーザーを10人集める。彼らの評価をもとに、自分がレビューした映画以外のレビューを予測する。評価が高いものTop10を最終的に出力する。

````python
# XとYに分ける
X = df_test_new[title].values 
Y = df_test_new.drop(title, axis = 1).values

# 学習用データ
x_train = X[0:610]
y_train = Y[0:610]


# テストデータ
x_test = X[610]

# モデル構築
knn10 = neighbors.KNeighborsRegressor(100, weights="distance", metric="euclidean")
model10 = knn10.fit(x_train, y_train)

# 予測
pred = model10.predict([x_test])
pred
````

レコメンド結果の出力

````python
recommend = pd.DataFrame(pred, columns=df_test_new.drop(title, axis = 1).columns)
recommend.sort_values(by = 0, axis=1, ascending=False, inplace=True) #降順に並び替える
recommend
````

上位10作品を出してみる。

````python
for i in range(0, 10):
    print(i+1,",", recommend.columns[i], "point: ", round(recommend.iloc[0, i], 2) )
````

今までの工程をまとめると、、

````python
def recommendation(data):
    # テストデータを訓練データと結合
    df_test_new = pd.concat([df_pivot, data], axis = 0, sort=False)
    # 欠損値は0として扱う
    df_test_new = df_test_new.fillna(0)
    # XとYに分ける
    X = df_test_new[title].values 
    Y = df_test_new.drop(title, axis = 1).values

    # 学習用データ
    x_train = X[0:610]
    y_train = Y[0:610]


    # テストデータ
    x_test = X[610]

    # モデル構築
    knn10 = neighbors.KNeighborsRegressor(500, weights="distance", metric="euclidean")
    model10 = knn10.fit(x_train, y_train)

    # 予測
    pred = model10.predict([x_test])
    # 結果をデータフレーム化
    recommend = pd.DataFrame(pred, columns=df_test_new.drop(title, axis = 1).columns)
    recommend.sort_values(by = 0, axis=1, ascending=False, inplace=True) #降順に並び替える
    # 出力
    for i in range(0, 10):
        print(i+1,",", recommend.columns[i], "point: ", round(recommend.iloc[0, i], 2) )
````

今まで観てきた映画とその評価を入力すると、レコメンドが出てくるシステムっぽいやつが出来ます。データの入力の段階がイケてないけど、多分ここからはWeb開発の話やUI・UXの話になってくると思います、、。レコメンドシステムの背後には、こういうシステムが回っているんだなぁと実感してもらえればありがたいです！

````python
##### ここから
title = ['Titanic (1997)', 'Jaws: The Revenge (1987)',
         'Harry Potter and the Prisoner of Azkaban (2004)', 'Notebook, The (2004)', 
         'Home Alone (1990)','Aladdin (1992)',
         'Ilsa, She Wolf of the SS (1974)' 'Wolf of Wall Street, The (2013)', "Intouchables (2011)",
         'About Time (2013)', 'Léon: The Professional (a.k.a. The Professional) (Léon) (1994)']
score = [[3.6, 2.2, 3.0, 3.8, 4.0, 3.7, 5, 4.6, 4.5, 5.0]]
##### ここまで入力してください！

index = ["MY REVIEW"]
df_test = pd.DataFrame(data = score, index = index ,columns = title)
recommendation(df_test)
````
